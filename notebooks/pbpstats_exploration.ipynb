```python
import pandas as pd

from pbpstats.client import Client


SETTINGS = {
    "dir": "/home/jupyter/nba/notebooks/data/pbpstats",
    "Boxscore": {"source": "file", "data_provider": "stats_nba"},
    "EnhancedPbp": {"source": "file", "data_provider": "stats_nba"},
    "Games": {"source": "file", "data_provider": "stats_nba"},
    "Pbp": {"source": "file", "data_provider": "stats_nba"},
    "Possessions": {"source": "file", "data_provider": "stats_nba"},
    "Shots": {"source": "file", "data_provider": "stats_nba"},
}
client = Client(SETTINGS)
```

```python
season = client.Season("nba", "2019-20", "Regular Season")
season_data = pd.DataFrame(season.games.final_games)
season_data
```

```python

game_id = season.games.final_games[-1]['game_id']
game = client.Game(game_id)
boxscore = game.boxscore
teams_boxscore_data = pd.DataFrame(boxscore.team_items)
teams_boxscore_data
```

```python
players_boxscore_data = pd.DataFrame(boxscore.player_items)
players_boxscore_data
```

```python
shots = game.shots
shots_data = pd.DataFrame(shots.data)
shots_data
```

```python
possessions = game.possessions
team_stats_data = pd.DataFrame(possessions.team_stats)
team_stats_data
```

```python
player_stats_data = pd.DataFrame(possessions.player_stats)
player_stats_data
```

```python
lineup_stats_data = pd.DataFrame(possessions.lineup_stats)
lineup_stats_data
```

```python
lineup_opponent_stats_data = pd.DataFrame(possessions.lineup_opponent_stats)
lineup_opponent_stats_data
```

```python
possession = possessions.items[0]
possession_stats_data = pd.DataFrame(possession.possession_stats)
possession_stats_data
```

```python
possession_data = possession.data
possession_data
```

```python
event = possession.events[-1]
event_stats_data = pd.DataFrame(event.event_stats)
event_stats_data
```

```python
event_base_stats_data = pd.DataFrame(event.base_stats)
event_base_stats_data
```

```python
event_data = event.data
event_data
```

```python
event_shot_data = event.shot_data
event_shot_data
```

```python
pbp = game.pbp
pbp_data = pd.DataFrame(pbp.data)
pbp_data
```

```python
pbp_item = pbp.items[0]
pbp_item_data = pbp_item.data
pbp_item_data
```

```python
enhanced_pbp = game.enhanced_pbp
enhanced_pbp_data = pd.DataFrame(enhanced_pbp.data)
enhanced_pbp_data
```
